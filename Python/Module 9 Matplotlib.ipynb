Sales_Volume where you need to add the following values based on total sales by category:/

“High Sales, High Quantity” where Quantity Sold >= 100 and Total Sales >$500 within a given month,/

“High Sales, Low Quantity” where Quantity Sold =<100 and Total Sales >= $500 within a given month,/

“Low Sales, Low Quantity” = Quantity Sold =< 100 and Total Sales =< $500 within a given month,/

“Low Sales, High Quantity" = Quantity Sold >= 100 and Total Sales =<$500 within a given month

```python
# Option 1: Case statement - Pandasql - sqldf()
# Option 2: If statement
# Option 3: Else-If statement
# Option 4: Loop
# Option 5: Map()

sv = product_sales[['category_name', 'quantity', 'total_sales', 'order_month_yr', 'order_year','order_month']]
sv

sv2 = sv.groupby(['category_name','order_month_yr'])['total_sales','quantity'].sum().reset_index()
sv_yr = sv.groupby(['category_name','order_year'])['total_sales','quantity'].sum().reset_index()
sv_mo_yr = sv.groupby(['category_name','order_year','order_month'])['total_sales','quantity'].sum().reset_index()
sv2
sv_mo_yr
```

```python
def sales_cat(row):
    if row ['total_sales'] >= 500 and row['quantity'] >= 100:
        return 'High Sales, High Quantity'
    elif row ['total_sales'] < 500 and row['quantity'] >= 100:
        return 'Low Sales, High Quantity'
    elif row ['total_sales'] >= 500 and row['quantity'] < 100:
        return 'High Sales, Low Quantity'
    else:
        return 'Low Sales, Low Quantity'
    
sv['sales_volume'] = sv.apply(sales_cat, axis = 1)
sv.head()
```
