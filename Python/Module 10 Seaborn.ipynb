```python
import pandas as pd
import numpy as np
product_sales=pd.read_csv("product_sales.csv")
```

```python
product_sales.head()
```

```python
product_sales = product_sales.drop('Unnamed: 0', axis=1)
product_sales.head()
```

```python
product_sales.columns
```

```python
#Minimum value
print('Total Sales Minimum:', product_sales['total_sales'].min())
#Maximum value
print('Total Sales Maximum:', product_sales['total_sales'].max())
#Mean value
print('Total Sales Mean:', product_sales['total_sales'].mean())
#median value
print('Total Sales Median:', product_sales['total_sales'].median())
#Standard Diviation value
print('Total Sales Standard Diviation:', product_sales['total_sales'].std())
#Mode value
print('Total Sales Mode:', product_sales['total_sales'].mode())
#variance value
print('Total Sales variance:', product_sales['total_sales'].var())
#Standard error of the mean value
print('Total Sales Standard error of the mean:', product_sales['total_sales'].sem())
#Rank value
print('Total Sales Rank:', product_sales['total_sales'].rank())
```

```python
import seaborn as sns
import matplotlib.pylab as plt

ps = product_sales['total_sales'].head(20)
ps = pd.DataFrame(ps)
ps
```

```python
sns.histplot(data=ps, x='total_sales', bins = 5, kde = True, color = 'red')
#Add lables to histogram
plt.xlabel('total_sales')
plt.ylabel('Frequency')
plt.title('Product Sales')

#Show plot
plt.show()
```

```python
sns.kdeplot(ps['total_sales'])
```

```python
mean = ps['total_sales'].mean()
st_dev = ps['total_sales'].std()

sns.kdeplot(ps['total_sales'])
plt.axvline(mean, color='purple', label= 'mean', linestyle='--')
plt.legend()
plt.savefig('total_sales_hist.jpg', dpi = 300, bbox_inches = 'tight')
plt.show()
```
