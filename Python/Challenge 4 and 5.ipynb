```python
#challenge 5
iris = pd.read_csv('iris_dirty.csv')
iris.head(10)
```

```python
iris3 = iris.drop(columns=['petal.width'])
iris3
```

```python
setosa = iris.query("variety == 'Setosa'")
print(setosa.head(10))

setosa_avg_petal = setosa['petal.width'].mean()
setosa_avg_petal
iris ['petal.width'] = iris['petal.width'].fillna(setosa_avg_petal) 
iris.head(10)
```

```python
#Find maximum sepal_width by variety using SQL
max_variety = sqldf("""Select variety,max([sepal.width]) as max_sepal From iris group by variety""")
max_variety
```

```python
max_variety2 = iris.pivot_table(columns = 'variety', values='sepal.width', aggfunc='max')
max_variety2
```
