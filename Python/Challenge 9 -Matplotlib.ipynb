```python
#Create a pivot data frame that shows the average petal width per species. 
from pandasql import sqldf
#pysqldf =lambda q: sqldf(q,globals())

avg_species = sqldf("""select species, 
avg([petal_width]) as avg_width
from iris group by species;""")
avg_species
```

```python
#Write code to group by species and find the median of all quantitative columns for each species. 
#Based on these results, which species has the shortest median bill depth?
ps = penguins.groupby(['species'])['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g'].median().reset_index()
ps
```

```python
mnb = ps.groupby(['species'])['bill_depth_mm'].min().reset_index()
mnb
```
