```python
#import libaries
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
dr = pd.read_csv('Warmup_DVD_Rental_3_Genres.csv')
dr.head(10)
```

```python
sns.histplot(dr['genre'])
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=dr, x='replacement_cost', hue='genre')
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=dr, x='genre', y='replacement_cost')
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=dr, x='genre', y='replacement_cost', hue='rating')
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=dr, x='genre', y='replacement_cost', hue='rating', inner="quart")
```
